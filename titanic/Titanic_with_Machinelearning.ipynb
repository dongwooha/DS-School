```python
import pandas as pd
```

### Load Dataset

```python
train = pd.read_csv("train.csv", index_col = "PassengerId")

print(train.shape)
train.head()
```

```python
test = pd.read_csv("test.csv", index_col = "PassengerId")

print(test.shape)
test.head()
```

### Preprocessing

### Encode Sex

```python
train["Sex_encode"] = train["Sex"].replace("male", 0).replace("female", 1)

print(train.shape)
train[["Sex", "Sex_encode"]].head()
```

```python
test["Sex_encode"] = test["Sex"].replace("male", 0).replace("female", 1)

print(test.shape)
test[["Sex", "Sex_encode"]].head()
```

### Fill in Missing Fare

```python
test["Fare"] = test["Fare"].fillna(0)

test[test["Fare"].isnull()]
```

```python
# C == 0, S == 1, Q == 2
# 2 * S == Q?
# S + S == Q?
# 문자를 인코딩할 경우, 경우의 수가 3가지가 넘아가면 One Hot Encoding 을 해줘야 함.

# One Hot Encoding
# Ture == 1, False == 0
# C == [Ture, False, False]
# S == [0, 1, 0]
# Q == [0, 0, 1]

train["Embarked_C"] = train["Embarked"] == "C"
train["Embarked_S"] = train["Embarked"] == "S"
train["Embarked_Q"] = train["Embarked"] == "Q"

train[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

```python
test["Embarked_C"] = test["Embarked"] == "C"
test["Embarked_S"] = test["Embarked"] == "S"
test["Embarked_Q"] = test["Embarked"] == "Q"

test[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

```python
train.loc[train["Age"] > 0, "Age_info"] = 1
train.loc[train["Age"].isnull(), "Age_info"] = 0

train[["Age", "Age_info"]].head()
```

```python
test.loc[test["Age"] > 0, "Age_info"] = 1
test.loc[test["Age"].isnull(), "Age_info"] = 0

test[["Age", "Age_info"]].head()
```

```python
train["Family_size"] = train["SibSp"] + train["Parch"]

train[["SibSp", "Parch", "Family_size"]]
```

```python
test["Family_size"] = test["SibSp"] + test["Parch"]

test[["SibSp", "Parch", "Family_size"]]
```

```python
mean_age = train["Age"].mean()
train["Age"] = train["Age"].fillna(mean_age)

train.loc[train["Age"] < 15, "Age_young"] = True
train.loc[(train["Age"] >= 15) & (train["Age"] < 50), "Age_middle"] = True
train.loc[train["Age"] >= 50, "Age_old"] = True

train["Age_young"] = train["Age_young"].fillna(False)
train["Age_middle"] = train["Age_middle"].fillna(False) 
train["Age_old"] = train["Age_old"].fillna(False) 

train[["Age", "Age_young", "Age_middle", "Age_old"]].head()
```

```python
test["Age"] = test["Age"].fillna(mean_age)

test.loc[test["Age"] < 15, "Age_young"] = True
test.loc[(test["Age"] >= 15) & (test["Age"] < 50), "Age_middle"] = True
test.loc[test["Age"] >= 50, "Age_old"] = True

test["Age_young"] = test["Age_young"].fillna(False)
test["Age_middle"] = test["Age_middle"].fillna(False) 
test["Age_old"] = test["Age_old"].fillna(False) 

test[["Age", "Age_young", "Age_middle", "Age_old"]].head()
```

### Train

```python
# Feature - Pclass, Sex, Fare, etc
# Label - Survived
```

```python
# Age, SibSp, Parch
# Age 구역
# SibSp, Parch, Fmaily size; solo, neclear, big

feature_name = ["Pclass", "Sex_encode", "Fare", "Embarked_C", "Embarked_S", "Embarked_Q", "Family_size", "Age_young", "Age_middle", "Age_old"]
feature_name
```

```python
X_train = train[feature_name]

print(X_train.shape)
X_train.head()
```

```python
X_test = test[feature_name]

print(X_test.shape)
X_test.head()
```

```python
label_name = "Survived"

y_train = train[label_name]

print(y_train.shape)
y_train.head()
```

### Use Decision Tree

```python
# scikit-learn = sklearn
from sklearn.tree import DecisionTreeClassifier

## 추천
model = DecisionTreeClassifier(max_depth = 10)
model
```

```python
# 1) fit (train) -> 2) predict (test)
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)

print(prediction.shape)
prediction[0:5]
```

### Submit

```python
submit = pd.read_csv("gender_submission.csv", index_col = "PassengerId")

submit["Survived"] = prediction

print(submit.shape)
submit.head()
```

```python
submit.to_csv("decision-tree_family_age_depth10.csv")
```
