### 1. 할인 가격

기름값을 계산하라. 많이 사면 할인을 해준다.

* 처음 20갤론: 갤론 당 90센트
* 다음 40갤로: 갤론 당 80센트
* 나머지: 갤론 당 75센트

예를 들어 35갤론을 구입했다면, 가격은 30달러(20 * 90 + 15 * 80)

```python
20 * 90 + 15 * 80
```

```python
def price_cal(gallon):

    if gallon <= 20:
        price = 90 * gallon
    elif gallon <= 40:
        price = 90 * 20 + 80 * (gallon-20)
    elif gallon > 40:
        price = 90 * 20 + 80 * 20 + 75 * (gallon-40)
    
    print('The price of ', gallon, "gallon is ", price)
    
    return(price)

price_cal(10)
price_cal(40)
price_cal(100)
```
